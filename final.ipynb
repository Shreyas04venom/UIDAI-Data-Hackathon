```python
from sklearn.ensemble import IsolationForest

# Prepare data
X = daily[['age_18_greater']]

# Train model
model = IsolationForest(contamination=0.05, random_state=42)
daily['anomaly'] = model.fit_predict(X)

# -1 means anomaly
anomalies = daily[daily['anomaly'] == -1]

print("Anomalous dates:")
display(anomalies)
```

```python
# Create risk score
daily['risk_score'] = (
    (daily['age_18_greater'].max() - daily['age_18_greater']) +
    (daily['anomaly'] == -1)*10000
)

# Rank risky dates
risk_table = daily.sort_values(by='risk_score', ascending=False)

risk_table.head(10)
```

```python
import folium
import pandas as pd
from sklearn.ensemble import IsolationForest
import glob

# -------------------- DATA PIPELINE --------------------

# Step 1: Discover CSV files
all_csv_files = glob.glob('/content/sample_data/**/*.csv', recursive=True)

# Step 2: Filter Aadhaar enrolment files
enrol_files = [f for f in all_csv_files if 'aadhar_enrolment' in f]

# Step 3: Load & merge
enrol_list = [pd.read_csv(f) for f in enrol_files]
enrol = pd.concat(enrol_list, ignore_index=True)

# Step 4: Date cleaning
enrol['date'] = pd.to_datetime(enrol['date'], format='%d-%m-%Y')
enrol = enrol.sort_values(by='date')

# Step 5: Daily totals
daily = enrol.groupby('date').sum(numeric_only=True)

# Step 6: Anomaly detection
X = daily[['age_18_greater']]
model = IsolationForest(contamination=0.05, random_state=42)
daily['anomaly'] = model.fit_predict(X)

# Step 7: Risk score
daily['risk_score'] = (
    (daily['age_18_greater'].max() - daily['age_18_greater']) +
    (daily['anomaly'] == -1)*10000
)

# Merge back
daily_reset = daily.reset_index()
enrol = pd.merge(enrol, daily_reset[['date','risk_score']], on='date', how='left')

# -------------------- STATE RISK --------------------

state_risk = enrol.groupby('state', as_index=False)['risk_score'].mean()

state_coords = {
    "Andaman & Nicobar Islands":[11.7401,92.6586],
    "Jammu & Kashmir":[33.7782,76.5762],
    "West Bengal":[22.9868,87.8550],
    "Odisha":[20.9517,85.0985],
    "Andhra Pradesh":[15.9129,79.7400],
    "Daman & Diu":[20.4283,72.8397],
    "Dadra & Nagar Haveli":[20.1809,73.0169],
    "Bihar":[25.0961,85.3131],
    "Uttar Pradesh":[26.8467,80.9462],
    "Assam":[26.2006,92.9376],
    "Meghalaya":[25.4670,91.3662],
    "Rajasthan":[27.0238,74.2179],
    "Maharashtra":[19.7515,75.7139],
    "Gujarat":[22.2587,71.1924],
    "Karnataka":[15.3173,75.7139],
    "Madhya Pradesh":[22.9734,78.6569],
    "Tamil Nadu":[11.1271,78.6569],
    "Kerala":[10.8505,76.2711],
    "Punjab":[31.1471,75.3412],
    "Haryana":[29.0588,76.0856],
    "Delhi":[28.7041,77.1025]
}

# -------------------- FILTER LOGIC --------------------

q99 = daily['risk_score'].quantile(0.52)
q97 = daily['risk_score'].quantile(0.48)

def risk_level(r):
    if r > q99:
        return "EXTREME","darkred"
    elif r > q97:
        return "HIGH","orange"
    else:
        return "NORMAL","green"

# -------------------- MAP --------------------

india_map = folium.Map(location=[22,78], zoom_start=5)

fg_extreme = folium.FeatureGroup(name=" EXTREME")
fg_high = folium.FeatureGroup(name=" HIGH")
fg_normal = folium.FeatureGroup(name=" NORMAL")

for _,row in state_risk.iterrows():
    state = row['state']
    risk = int(row['risk_score'])

    if state in state_coords:
        lat,lon = state_coords[state]
        level,color = risk_level(risk)

        marker = folium.CircleMarker(
            location=[lat,lon],
            radius=risk/10000,
            popup=f"<b>{state}</b><br>Risk:{risk}<br>Level:{level}",
            tooltip=f"{state} | {level}",
            color=color,
            fill=True,
            fill_opacity=0.7
        )

        if level=="EXTREME":
            marker.add_to(fg_extreme)
        elif level=="HIGH":
            marker.add_to(fg_high)
        else:
            marker.add_to(fg_normal)

fg_extreme.add_to(india_map)
fg_high.add_to(india_map)
fg_normal.add_to(india_map)

folium.LayerControl(collapsed=False).add_to(india_map)

india_map.save("risk_map.html")
india_map
```
